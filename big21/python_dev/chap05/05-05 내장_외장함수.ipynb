```python
# 내장 함수
# filter(F, Iterable)
# 요건 : 리스트에서 양수 추출

# filter 사용x
datas = [1,-3,2,0,-5,6]
result = []     # 결과 저장할 빈 리스트 생성
for data in datas:
    if data > 0 :
        result.append(data)
print(result)
```

```python
# filter(조건이 들어있는 함수, 리스트) 사용 코드로 변경 - return 값은 T/F로 나와야 함.
def positive(x):    # x는 단일값
    return x > 0         # boolean(True, False) : 비교연산자
```

```python
print(list(filter(positive, datas))) # Iterable filter object. Iterable한 객체들은 리스트 변환 가능
```

```python
# lambda 명령: 함수 오브젝트 생성
# 기본 구조 -> lambda 매개변수 : 실행 코드
# positive -> lambda x : x > 0
list(filter(lambda x:x>0, datas))
```

```python
# (lambda a,b:a+b)(3,4)
aaa = lambda a,b:a+b    # 보통 이렇게 안 씀. 이럴 거면 def하는 게..
aaa(3,4)
```

```python
# map(function, iterable) : function은 가공
# 요건: 리스트 값을 2배 처리
datas = [1,2,3,4]   # -> [2,4,6,8]
result = []         # 결과 저장할 빈 리스트 생성
for data in datas:
    result.append(data*2)
print(result)
```

```python
# map 사용
def two_times(x):   # x: 단일값
    return x*2

print(list(map(two_times, datas)))
```

```python
list(map(lambda x:x*2, datas))
```

```python
def plus_one(x):
    return x+1
```

```python
# 2배
print(list(map(two_times, datas)))
print(list(map(plus_one, datas)))
```
