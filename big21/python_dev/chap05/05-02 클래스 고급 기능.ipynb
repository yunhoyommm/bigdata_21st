```python
# 요건
# 1. 박씨 관리하는 집: 성이 "박" (클래스변수)
# 2. fullname: 성+이름
# 3. 여행 메소드: 장소를 입력하면 "성+이름, 장소 여행을 가다"

class HousePark:
    lastName = "박"

    def __init__(self, name):
        self.fullname = self.lastName + name

    # def setName(self, name):
    #     self.fullname = self.lastName + name
    
    def travel(self, where):
        print(f"{self.fullname}, {where} 여행을 가다")
```

```python
aa = HousePark("응용")

```

```python
print(aa.lastName)
# aa.setName("응용")
print(aa.fullname)
aa.travel("부산")
```

```python
# 클래스의 상속
# HousePark 을 상속받는 HouseKim
class HouseKim(HousePark):
    lastName = "김"
                        # HouseKim에 __init__ 이 없으니 HousePark으로 가서 init 찾음
    # 김줄리엣, 독도 여행을 가다 => 김줄리엣은 독도로 여행합니다
    def travel(self, where):
        print(f"{self.fullname}은 {where}로 여행합니다.")
```

```python
juliet = HouseKim("줄리엣")
```

```python
juliet.travel("독도")

```

```python
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        return "동물이 소리를 냅니다."
    
class Dog(Animal):
    def speak(self):
        return f"멍멍! 저는 {self.name}입니다."

class Cat(Animal):
    def speak(self):
        return f"야옹! 저는 {self.name}입니다."
```

```python
ant = Animal("개미")
dog = Dog("구름이")
cat = Cat("나비")


print(ant.speak())
print(dog.speak())
print(cat.speak())
```

```python
class Employee:
    def __init__(self, name, base_salary):
        self.name=name
        self.base_salary=base_salary
    
    def calculate_salary(self):
        return self.base_salary
    
    def info(self):
        return f"{self.name}의 급여는 {self.calculate_salary()}원입니다."
    
class Manager(Employee):
    def __init__(self, name, base_salary, bonus):
        super().__init__(name, base_salary)
        self.bonus = bonus

    def calculate_salary(self):
        return self.base_salary+self.bonus

class Developer(Employee):
    def __init__(self, name, base_salary, overtime_pay):
        super().__init__(name, base_salary)
        self.overtime_pay = overtime_pay

    def calculate_salary(self):
        return self.base_salary+self.overtime_pay
    
```

```python
a = Employee("홍길동", 1000)
b = Manager("고길동", 1000, 5000)
c = Developer("마이콜", 1000, 9000)

print(a.info())
print(b.info())

print(a.calculate_salary())

print(b.calculate_salary())
print(c.calculate_salary())
```
