```python
# 예외 처리
# 먼저 코드 작성 후 예외 확인 후에 except에서의 예외 선정

try:
    a = [1,2]
    print(a[3])
    # 4/0
except IndexError:
    print("인덱싱 할 수 없습니다")

print("정상 종료")
```

```python
try:
    a = [1,2]
    print(a[3])
except IndexError as e:     # as e 는 꺼내 쓰는 거까지 포함.
    # print("인덱싱할 수 없습니다.")
    print(e) # print함수가 print(e.__str__()) 변경해서 실행(java에서는 toString)
```
