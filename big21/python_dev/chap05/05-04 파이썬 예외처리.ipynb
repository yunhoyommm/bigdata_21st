```python
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def speak(self):    # 상속받는 클래스는 speak 메소드를 반드시 오버라이딩 하게 만든다.
        pass

class Dog(Animal):      # Animal 상속받았으니 Dog 클래스에 speak 메소드 없으면 에러
    def speak(self):
        return "멍멍"

class Cat(Animal):
    def speak(self):
        return "야옹"
    
class Cow(Animal):
    def speak(self):
        return "음메"
    
class Duck(Animal):         # TypeError: Can't instantiate abstract class Duck without an implementation for abstract method 'speak'
    def speak(self):        # 위 에러 보고 speak 생성해야 함을 인지 후 수정("speakDuck->speak")
        return "꽥"

dog = Dog()
print(dog.speak())

cat = Cat()
print(cat.speak())

cow = Cow()
print(cow.speak())

duck = Duck()

animals = [Dog(), Cat(), Cow(), Duck()]

for animal in animals:
    print(animal.speak())
```

```python
class Bird:         # 모든 새는 난다. but 비행 방법은 다름. -> 다형성으로 구현
    def fly(self):  # 이 메소드 없으면 덕타이핑으로 구현. 
        raise NotImplementedError("추상적이다.")   # 예외 object 생성해서 던짐.
        # 추상적이고 다른 메서드로 자식 fly에서 구현되니 이 메소드가 구현되면 에러 던짐.
        
```

```python
b = Bird()  # object 생성, fly 호출해서 실행
try:
    b.fly()
except NotImplementedError as e:
    print(e)
```

```python
class Eagle(Bird):
    def fly(self):
        print("very fast")

eagle = Eagle()

print(type(eagle))
eagle.fly()
```

```python
# 오류 처리하는 클래스
class MyError(Exception):  # Exception의 오류 처리하는 변수, 메서드 사용
    # 여기서 에러 만드는 거니
    # 오류 발생 원인을 저장, 출력해야 함.
    # 저장: 변수 선언
    def __init__(self):
        self.message = "허용되지 않는 별명입니다."
    # 문자열 출력: 메소드
    def __str__(self):  # 출력하고자 하는 변수 반환
        return self.message
```

```python
def say_nick(nick):     # nick을 받아 처리하는 함수, nick == "바보" 에러 발생시키고, 아니면 출력
    if nick == "바보":
        raise MyError()
    print(nick)
```

```python
say_nick("천사")

try:
    say_nick("바보")
except MyError as e:
    print(e)
```
