```python
# 클래스 : 비슷한 성격의 데이터(변수)와 기능(메소드)을 하나로 묶어,
#         오브젝트를 찍어내는 설계도(틀)
# 설계도대로 생성 -> 객체 : 오브젝트, 인스턴스
# 반복 사용 의미가 있다 : 함수 비슷하다

# 왜 사용하는가? 
# 예를 들어 학생 1명의 정보 저장 : 10개의 변수가 필요(이름, 나이, 국어점수, ...)
# 1000명이라면 각각 저장하면 10000개의 선언이 필요.
# 10개 공간을 만드는 객체를 생성하면 1000번만 생성해도 됨.
```

```python
# 한 명의 학생 정보를 저장하는 클래스 선언
class Student:  # Student라는 클래스를 선언해라. 이름, 점수 저장하는 공간 선언
    def __init__(self, name, score):    # __init__() : 생성될 때 자동 호출되는 메소드(생성자)
        self.name = name        # self.name : 오브젝트 안에 name이라는 변수 선언하고 name값을 대입
        self.score = score      
    def getStudent(self):       # 주소값이 self에 들어감. self가 있어야 오브젝트 접근 가능
        print(f"{self.name} : {self.score}")
```

```python

student1 = Student("홍길동", 90)        # 결국 student1은 self와 같음.
# 클래스에 () 붙으면 오브젝트 생성 명령(주소값 만들어짐) -> __init__ 초기화하면서 self에 주소값 들어감.

# student2 = Student("고길동", 80)    
# student3 = Student("마이콜", 70)
student1.getStudent()
```

```python
# 클래스를 쓰는 이유
# 1. 데이터(변수)와 기능(메소드)을 묶기
# 2. 재사용: 설계도 하나로 객체 무한 생성 가능
# 3. 확장성: 상속으로 기존 코드를 물려받아 확장 가능
# 4. 현실 모델링: 현실의 사물을 자연스럽게 표현 가능 (예. 자동차, 사물, 스마트폰)
```

```python
# 계산기 클래스
class Calculator:
    def __init__(self):
        self.result = 0
    def adder(self, num):
        self.result += num  # result에 num 누적
        return self.result
    def sub(self, num):
        self.result -= num
        return self.result
```

```python
# 오브젝트 2개 생성
cal1 = Calculator()
cal2 = Calculator()
```

```python
print(cal1.adder(3))
print(cal1.sub(2))
```

```python
# 지금부터 계산기 클래스 생성
# 요건: 더하기 기능, 결과 반환 계산
# 1. 클래스 이름: Calculator, 클래스 이름은 해당 클래스의 용도를 설명할 수 있는 이름으로 작성
# 2. 변수: 결과 저장, 반환. 처음에 0
# 3. 기능(메소드): 더하기 메소드. 더할 값을 받아와야 함. 더한 결과를 result 저장. 

class Calculator:                   # class Calculator(object)와 동일
    def __init__(self):             # object class의 __init__을 재정의(overriding)
        self.result = 0             # 결과 저장 변수가 처음에 0으로 초기화

    def adder(self, number):        # result + number -> return result
        self.result += number
        return self.result
```

```python
# 요건 : 더하기 기능을 하고, 결과 반환 계산기
class Calculator:
    def __init__(self):
        self.result = 0
    
    def adder(self, number):
        self.result += number
        return self.result
```

```python
cal1 = Calculator()
print(cal1.adder(1))
print(cal1.adder(9))
print(cal1.adder(-9))

print(cal1.result)
```

```python
# 학생 정보 저장하고 출력하는 클래스
# 학생 정보: 학번, 학과, 이름, 성별
# 출력: 학번, 학과, 이름

class Student:
    def __init__(self, studentNumber, department, name, gender):
        self.studentNumber = studentNumber
        self.department = department
        self.name = name
        self.gender = gender

    def printStudent(self):
        print("학번:", self.studentNumber, ", 학과:", self.department, ", 이름:", self.name)
        return None

    def printStudentDict(self):
        print = {
            "studentNumber":self.studentNumber,
            "department":self.department,
            "name":self.name,
            "gender":self.gender
        }
        return print
```

```python
stu1 = Student(1,"컴퓨터공학과","홍길동","남자")
stu1.printStudent()

stu1dict = stu1.printStudentDict()
print(f"학번: {stu1dict["studentNumber"]}, 학과: {stu1dict.get("department")}, 이름: {stu1dict.get("name")}")
```
