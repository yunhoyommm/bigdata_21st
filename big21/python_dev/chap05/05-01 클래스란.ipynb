```python
# 클래스 : 비슷한 성격의 데이터(변수)와 기능(메소드)을 하나로 묶어,
#         오브젝트를 찍어내는 설계도(틀)
# 설계도대로 생성 -> 객체 : 오브젝트, 인스턴스
# 반복 사용 의미가 있다 : 함수 비슷하다

# 왜 사용하는가? 
# 예를 들어 학생 1명의 정보 저장 : 10개의 변수가 필요(이름, 나이, 국어점수, ...)
# 1000명이라면 각각 저장하면 10000개의 선언이 필요.
# 10개 공간을 만드는 객체를 생성하면 1000번만 생성해도 됨.
```

```python
# 한 명의 학생 정보를 저장하는 클래스 선언
class Student:  # Student라는 클래스를 선언해라. 이름, 점수 저장하는 공간 선언
    def __init__(self, name, score):    # __init__() : 생성될 때 자동 호출되는 메소드(생성자)
        self.name = name        # self.name : 오브젝트 안에 name이라는 변수 선언하고 name값을 대입
        self.score = score      
    def getStudent(self):       # 주소값이 self에 들어감. self가 있어야 오브젝트 접근 가능
        print(f"{self.name} : {self.score}")
```

```python

student1 = Student("홍길동", 90)        # 결국 student1은 self와 같음.
# 클래스에 () 붙으면 오브젝트 생성 명령(주소값 만들어짐) -> __init__ 초기화하면서 self에 주소값 들어감.

# student2 = Student("고길동", 80)    
# student3 = Student("마이콜", 70)
student1.getStudent()
```

```python
# 클래스를 쓰는 이유
# 1. 데이터(변수)와 기능(메소드)을 묶기
# 2. 재사용: 설계도 하나로 객체 무한 생성 가능
# 3. 확장성: 상속으로 기존 코드를 물려받아 확장 가능
# 4. 현실 모델링: 현실의 사물을 자연스럽게 표현 가능 (예. 자동차, 사물, 스마트폰)
```

```python
# 계산기 클래스
class Calculator:
    def __init__(self):
        self.result = 0
    def adder(self, num):
        self.result += num  # result에 num 누적
        return self.result
    def sub(self, num):
        self.result -= num
        return self.result
```

```python
# 오브젝트 2개 생성
cal1 = Calculator()
cal2 = Calculator()
```

```python
print(cal1.adder(3))
print(cal1.sub(2))
```

```python
# 지금부터 계산기 클래스 생성
# 요건: 더하기 기능, 결과 반환 계산
# 1. 클래스 이름: Calculator, 클래스 이름은 해당 클래스의 용도를 설명할 수 있는 이름으로 작성
# 2. 변수: 결과 저장, 반환. 처음에 0
# 3. 기능(메소드): 더하기 메소드. 더할 값을 받아와야 함. 더한 결과를 result 저장. 

class Calculator:                   # class Calculator(object)와 동일
    def __init__(self):             # object class의 __init__을 재정의(overriding)
        self.result = 0             # 결과 저장 변수가 처음에 0으로 초기화

    def adder(self, number):        # result + number -> return result
        self.result += number
        return self.result
```

```python
# 요건 : 더하기 기능을 하고, 결과 반환 계산기
class Calculator:
    def __init__(self):
        self.result = 0
    
    def adder(self, number):
        self.result += number
        return self.result
```

```python
cal1 = Calculator()
print(cal1.adder(1))
print(cal1.adder(9))
print(cal1.adder(-9))

print(cal1.result)
```

```python
# 학생 정보 저장하고 출력하는 클래스
# 학생 정보: 학번, 학과, 이름, 성별
# 출력: 학번, 학과, 이름

class Student:
    def __init__(self, studentNumber, department, name, gender):       # 2번. 파이썬 인터프리터가 실행. positional arguments라 인터프리터가 첫 번째 self에 주소값 넣어줌.
        self.studentNumber = studentNumber
        self.department = department
        self.name = name
        self.gender = gender

    def printStudent(self):
        print(f"학번: {self.studentNumber}, 학과: {self.department}, 이름: {self.name}")
        return None

    def studentDict(self):
        student = {
            "studentNumber":self.studentNumber,
            "department":self.department,
            "name":self.name,
            "gender":self.gender
        }
        return student
    
```

```python
stu1 = Student(1,"컴퓨터공학과","홍길동","남자")        # 1번
stu1.printStudent()                                  # stu1이 가지고 있는 주소값을 해당 메소드의 self에 넣어줌(인터프리터가)

stu1dict = stu1.studentDict()
print(f"학번: {stu1dict["studentNumber"]}, 학과: {stu1dict.get("department")}, 이름: {stu1dict.get("name")}")

Student.printStudent(stu1)                           # print_student(self) self에 stu1 주소값이 들어가므로 실행 가능.
```

```python
# 클래스 고급 기능
# 1. 클래스 변수: 클래스의 변수 -> 클래스 소유의 변수, 메소드 밖에서 만듦.
# - 용도: 모든 오브젝트가 공유하는 변수
# - 오브젝트 변수: self로 만든 변수

# 예 설명: 인터넷 정보 제공하는 클래스
class Service:                    # self가 없다 -> object에 뭘 만들 수 없다.
    secret = "영구는 외계인이다."   # 클래스 변수
```

```python
# 합을 구해주는 서비스를 추가
class Service:
    secret = "영구는 외계인이다"

    def __init__(self, name):
        self.name = name
    # self 가 필요 없다. 그러나 pey.sum(1,1) 호출될 때 self가 필요하다.
    # def sum(a, b):      # self는 필요없는데 self 줘야 함.
    #     result = a + b  # 지역변수
    #     print(f"{a} + {b} = {result} 입니다")
    def sum(self, a, b):
        result = a + b
        print(f"{self.name}님의 요청: {a} + {b} = {result} 입니다")
        return
```

```python
pey = Service("Pey")         
kim = Service("Kim")         
pey.sum(3,2)                    # pey의 self가 def sum의 self를 넣어줌. 이 때는 self 있어야
# print(Service.sum(3,2))       # 클래스로 접근하면 접근하는 메소드에 self 없어도 됨.
                                # self가 없어서 주석 풀면 Error(self에 3들어감.)
Service.sum(kim,3,4)
```

```python
# 오브젝트 변수
# __init__은 파이썬 인터프리터에 의해 1번만 호출. 우리가 호출할 수 없음
# self 받는 메소드에서만 오브젝트 변수 선언 가능
class Service:
    secret = "영구는 외계인이다."

    def __init__(self, name):
        self.name = name

    def setName(self, name):
        self.name = name        # 오브젝트에 name 변수 선언
    
    def sum(self, a, b):
        result = a + b
        print(f"{self.name}님 {a} + {b} = {result} 입니다.")
```

```python
pey = Service()
pey.setName("홍길동")   # 주석하면 Error. 그래서 __init__ 필요
pey.sum(1,1)
```

```python
pey = Service("hong")
print(pey.name)
pey.setName("홍길동")
print(pey.name)
pey.sum(1,1)
```
