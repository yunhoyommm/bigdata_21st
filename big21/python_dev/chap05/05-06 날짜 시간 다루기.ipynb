```python
import time

# 현재 시간을 가지고 와서 년월일시분초로 분리해 준다
# time.localtime(time.time())
dt = time.localtime()
print(dt)
print(dt.tm_year)   # %Y
print(dt.tm_mon)
print(dt.tm_mday)
```

```python
# datetime module : 날짜, 시간 다루는 모듈
# time보다 datetime 쓰는 게 나음.
# 클래스
#  - date: 날짜만 처리
#  - time: 시간만 처리
#  - datetime: 날짜+시간
#  - timedelta: 기간/간격 차이 구할 수 있는 클래스
#  - timezone / zoneinfo: 시간대, 한국 +09:00(KST)
```

```python
# 특정 날짜와 시간으로 생성
from datetime import date, time, datetime
d = date(2026,6,2)  # 날짜 오브젝트 생성
print(d)

t = time(15,44,0)   # 정수로 들어가야 함.
print(t)

# dt = datetime(2026,6,2,15,44,0) # 06은 문자열.
dt = datetime(2026,6,2) # 시,분,초 생략(00:00:00)
print(dt)
```

```python
# 현재 시각 가져오기(많이 사용)
from datetime import datetime, date, timezone

print(datetime.now())   # 현재 로컬(지역) 날짜 + 시간. 사용하는 컴퓨터의 날짜,시간
# print(datetime.now())  # 불안전. 로컬pc마다 초, 밀리초가 다름.
print(datetime.now(timezone.utc))   # 안전한 값

print(date.today())     # 오늘 현재 날짜
```

```python
# 3. 속성(클래스 안에 있는 변수) 꺼내기 
# 속성: get/set 할 수 있는 변수
dt = datetime(2026,6,2,15,30,0)         
print(dt)

print(dt.year, dt.month, dt.day)      # 직접 오브젝트의 속성에 접근 가능
print(dt.hour, dt.minute, dt.second)
print(dt.weekday())     # 요일이 숫자로 출력(월요일:0, 화요일:1, ... , 일요일:6)
print(dt.isoweekday())  # 요일이 숫자로 출력(월요일:1, 화요일:2, ... , 일요일:7)
```

```python
week = ["월","화","수","목","금","토","일"]
dt = datetime(2026,6,2)

#           2026-06-02 00:00:00
print(f"{dt:%Y-%m-%d}은 {week[dt.weekday()]}요일")
```

```python
# 문자열 <-> datetime object
# 1. datetime -> 문자열 : strftime()  (strftime()보다 f-string 더 많이 쓰긴 함.)
dt = datetime(2026,6,2,15,30,0)
print(dt.strftime("%Y-%m-%d"))
print(f"{dt:%Y-%m-%d}")     # 이 방식 더 자주 씀.
# 2. 날짜 시간형태로 된 문자열 -> datetime object : strptime() (중요. 대체가 안 됨.)
d_str = '2026/06/02 15:30'
dt = datetime.strptime(d_str, "%Y/%m/%d %H:%M")     # datetime object 생성
print(dt.year, dt.month, dt.day, sep="-")

# 년도 추출
print(d_str[:4])

print(d_str.split()[0].split("/")[0])   # 년
print(d_str.split()[0].split("/")[1])   # 월
```

```python
# 날짜 계산 : timedelta
from datetime import datetime, timedelta

now = datetime(2026,6,2,15,0)   # 2026-06-02 15시 0분 0초   2026-06-02 15:00:00
# 10일 뒤
print(now + timedelta(days=10))
# 3시간 전
print(now - timedelta(hours=3))
# 2주 뒤
print(now + timedelta(weeks=2))
# 복합 : 1일 6시간 30분 뒤
print(now + timedelta(days=1, hours=6, minutes=30))
```

```python
# 두 날짜,시간 object의 차이(빼기) => timedelta 반환

d1 = datetime(2026,6,2)
d2 = datetime(2026,12,31)
diff = d2 - d1
print(type(diff))
print(diff)
print(diff.days)
```

```python
# 비교와 정렬
d1 = datetime(2026,6,2,9)
d2 = datetime(2026,6,2,18)
print(d1 < d2)
print(d1 > d2)


dates = [datetime(2026,6,2), datetime(2026,6,1), datetime(2026,6,3)]
print(sorted(dates))
```

```python
# timezone 다루기
naive = datetime(2026,6,2,15,30)
# 어느 타임존 : +09:00(KST)
print(naive.tzinfo)
```

```python
# timezone 명시하면 어떤 이익이 있는지 확인
from datetime import datetime
from zoneinfo import ZoneInfo
# 한국 시간으로 명시
kst = datetime(2026,6,2,15,30, tzinfo=ZoneInfo("Asia/Seoul"))
print(kst)
print(kst.astimezone(ZoneInfo("America/New_York")))
```

```python
# 예제 1. 로그 타임스탬프 처리하기
# [2026-06-02 17:48:18] 크롤링 시작

from datetime import datetime, timezone
def log(message:str) -> None :
    # 밀리초까지, UTC 기준 일관되게 기록
    ts = datetime.now(timezone.utc).strftime("%Y-%m-%d %H:%M:%S.%f")[:-3]   # 밀리초까지 가져오니
    print(f"[{ts}] {message}")
```

```python
log("크롤링 시작")
```

```python
# 예제 2. d-day : 남은 날짜 계산
from datetime import date

def d_day(target:date) -> str:  # 날짜 입력해서 남은 날짜 문자열로 반환
    diff = (target - date.today()).days     # date - date = timedelta
    return f"D-{diff}"
```

```python
d_day(date(2026,11,19))
```

```python
# 예제 3. 특정 코드 실행 시간 계산
import time         # datetime 말고 time 쓰기(정확히 보장이 안 됨.)
start = time.perf_counter() # 현재 시간. perf_counter()가 안전한 접근

sum(range(1000000000))
end = time.perf_counter() - start 
print(f"{end:.4f} 초 소요")
```
