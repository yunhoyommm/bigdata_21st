```python
# 함수의 기본 구조
# def 함수명 (입력값): 입력값은 없을 수 있다.
#   반복될 코드
#   return 반환값 # return은 반드시 있어야 함.
```

```python
# 3개 상품 가격에 부가세 포함 가격 출력
price1 = 10000
print(f"부가세 포함 가격: {price1*1.1}")
price2 = 25000
print(f"부가세 포함 가격: {price2*1.1}")
price3 = 7000
print(f"부가세 포함 가격: {price3*1.1}")
```

```python
# 위 코드를 처리하는 함수 선언
def include_tax (price):
    print(f"부가세 포함 가격: {price*1.1}")
    return None                             # 반환값 없어도 return None(None은 조건문에서 쓰임.)
                                            # return 기능: 반환, 함수 종료
                                            # 반환값 없으면 return만 적어도o
                                            # return 자체를 안 적으면 인터프리터에서 만들어줌
```

```python
print(include_tax(10000))

include_tax(10000)
include_tax(25000)
include_tax(7000)
```

```python
# 함수 선언과 실행
# 두 정수를 입력받아 합을 구하는 함수(입력 -> 코드 -> 반환)
# 입력 : 두 정수
# 코드: 두 정수를 더하기
# 반환 : 합
def sum_(int_a, int_b):
    result = int_a + int_b
    return result

def sum__(int_a, int_b):
    return int_a + int_b

sum_(10,20), sum__(10,20)   # 튜플 형태로 출력

```

```python
# 용어: 매개변수, 인자 구분
# 매개변수: 받을 그릇.
# 인자: 대입(입력)되는 값
def multiple(x,y):  # x, y: 매개변수(parameter)
    return x*y

multiple(3,4)       # 3, 4: 인자(argument)
```

```python
# 함수의 매개변수 설명
# 인자 다루는 방식 2가지
# 1. 위치 기반 인자(Positional Arguments)
def divide(a,b):
    return a/b

print(divide(10,2))
print(divide(2,10))


# 2. 키워드 기반 인자(Keyword Arguments)
print(divide(a=10, b=2))
print(divide(b=2, a=10))
```

```python
# 참고만. 2. 입력값 x, 결과값 o 경우
a = 10

def aaa():
    a = 20
    return a
print(aaa())
print(a)
```

```python
# 매개변수의 기본값 : 자주 사용하는 값을 미리 정해둔 것. 호출 시 생략 가능
def greet(name, greeting="반갑습니다"):
    print(f"{greeting}, {name}님")
```

```python
# 3. 가변 인자(arguments): *(positional arguments)
# 2개 정수의 합을 구하는 함수 선언, 실행

def _sum2(x,y):
    total = x+y
    return total

print(_sum2(1,2))

# 3개 정수의 합을 구하는 함수 선언, 실행
def _sum3(x,y,z):            # 3개에 대한 _sum이 만들어져 _sum 2개짜리는 사라짐
    return x+y+z

print(_sum2(1,2))            # python은 오버로딩 없음. 같은 이름의 함수x
print(_sum3(1,2,3))
```

```python
# 가변인자 적용
def __sum(*numbers):    # *numbers : 가변 매개변수(들어올 때는 튜플로 처리)
    total = 0
    for number in numbers:
        total += number
    return total
```

```python
print(__sum(1,2))       # ,니까 tuple
print(__sum(1,2,3))
print(__sum(1,2,3,4))
```

```python
# 4. 가변 인자 : ** -> 개수가 정해지지 않은 키워드 인자(keyword arguments)
# 함수 내부에서 딕셔너리 형태로 처리
def profile(**info):    # ** 두 개이면 딕셔너리로 처리하겠다.
    for key, value in info.items():
        print(f"{key}, {value}")
```

```python
profile(name="홍길동", age=28, job="developer") 
# keyword arguments 몇 개든 모든 데이터를 받을 수 있음.
```

```python
# return 의 또 다른 쓰임새 : 반환, 함수 종료
# 나의 별명? 바보라고 하면 아무 말도 안 함.
def say_nick(nick):
    if nick == "바보" :
        return          # 함수 안에서 return 실행되면 함수 종료
    print(f"나의 별명은 {nick}입니다.")
```

```python
say_nick("바보")
say_nick("천사")
```
