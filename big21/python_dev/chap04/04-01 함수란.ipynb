```python
# 함수란: 반복 사용 코드를 묶어놓고 이름 부여해서 사용
# 특정 작업을 수행하는 코드 묶음(블록)
# 이름을 붙여 필요할 때마다 불러 쓸 수 있게 만든 것

# 커피 머신 
# 사람(드립): 원두 갈기 -> 물 끓이기 -> 추출하기 -> 컵에 따르기
# 커피머신
# 1. 사람이 원두, 물 넣기 : 입력값 넣는다(argument)
# 2. 머신 내부의 복잡한 추출과정: 함수 내부 코드 실행
# 3. 나오는 커피: 반환값(return value)
```

```python
# 함수 사용 목적
# 1. 코드의 재사용      *제일 중요한 목적
# 2. 중복 제거
# 3. 가독성 효과
# 4. 유지보수 좋아짐    *중요한 목적
# 원칙: Don't Repeat Yourself!
```

```python
# 함수의 기본 구조
# def 함수명 (입력값): 입력값은 없을 수 있다.
#   반복될 코드
#   return 반환값 # return은 반드시 있어야 함.
```

```python
# 3개 상품 가격에 부가세 포함 가격 출력
price1 = 10000
print(f"부가세 포함 가격: {price1*1.1}")
price2 = 25000
print(f"부가세 포함 가격: {price2*1.1}")
price3 = 7000
print(f"부가세 포함 가격: {price3*1.1}")
```

```python
# 위 코드를 처리하는 함수 선언
def includeTax (price):
    print(f"부가세 포함 가격: {price*1.1}")
    return None                             # 반환값 없어도 return None(None은 조건문에서 쓰임.)
                                            # return 기능: 반환, 함수 종료
                                            # 반환값 없으면 return만 적어도o
                                            # return 자체를 안 적으면 인터프리터에서 만들어줌
```

```python
print(includeTax(10000))

includeTax(10000)
includeTax(25000)
includeTax(7000)
```

```python
# 함수 선언과 실행
# 두 정수를 입력받아 합을 구하는 함수(입력 -> 코드 -> 반환)
# 입력 : 두 정수
# 코드: 두 정수를 더하기
# 반환 : 합
def sum_(int_a, int_b):
    result = int_a + int_b
    return result

def sum__(int_a, int_b):
    return int_a + int_b

sum_(10,20), sum__(10,20)   # 튜플 형태로 출력

```

```python
# 입력값과 결과값에 따른 함수의 형태
# 1. 입력값, 결과값 다 있는 경우
# def 함수명(입력값):
#   return 결과값

# 2. 입력값 x, 결과값 o 경우
# def 함수명():
#   return 결과값

# 3. 입력값 o, 결과값 x 경우
# def 함수명(입력값):
#   return none: 함수 선언 시 return 안 적어도 자동 생성

# 4. 입력값, 결과값 다 없는 경우
# def 함수명():
#   return none
```

```python
# 용어: 매개변수, 인자 구분
# 매개변수: 받을 그릇.
# 인자: 대입(입력)되는 값
def multiple(x,y):  # x, y: 매개변수
    return x*y

multiple(3,4)       # 3, 4: 인자
```

```python
# 인자 다루는 방식 2가지
# 1. 위치 기반 인자(Positional Arguments)
def divide(a,b):
    return a/b

print(divide(10,2))
print(divide(2,10))


# 2. 키워드 기반 인자(Keyword Arguments)
print(divide(a=10, b=2))
print(divide(b=2, a=10))
```

```python
# 참고만. 2. 입력값 x, 결과값 o 경우
a = 10

def aaa():
    a = 20
    return a
print(aaa())
print(a)
```

```python
# 매개변수의 기본값 : 자주 사용하는 값을 미리 정해둔 것. 호출 시 생략 가능
def greet(name, greeting="반갑습니다"):
    print(f"{greeting}, {name}님")
```

```python
greet("홍길동","안녕하세요")
greet("고길동")
greet("마이콜")
```
