```python
# 함수란: 반복 사용 코드를 묶어놓고 이름 부여해서 사용
# 특정 작업을 수행하는 코드 묶음(블록)
# 이름을 붙여 필요할 때마다 불러 쓸 수 있게 만든 것

# 커피 머신 
# 사람(드립): 원두 갈기 -> 물 끓이기 -> 추출하기 -> 컵에 따르기
# 커피머신
# 1. 사람이 원두, 물 넣기 : 입력값 넣는다(argument)
# 2. 머신 내부의 복잡한 추출과정: 함수 내부 코드 실행
# 3. 나오는 커피: 반환값(return value)
```

```python
# 함수 사용 목적
# 1. 코드의 재사용      *제일 중요한 목적
# 2. 중복 제거
# 3. 가독성 효과
# 4. 유지보수 좋아짐    *중요한 목적
# 원칙: Don't Repeat Yourself!
```

```python
# 함수의 기본 구조
# def 함수명 (입력값): 입력값은 없을 수 있다.
#   반복될 코드
#   return 반환값 # return은 반드시 있어야 함.
```

```python
# 3개 상품 가격에 부가세 포함 가격 출력
price1 = 10000
print(f"부가세 포함 가격: {price1*1.1}")
price2 = 25000
print(f"부가세 포함 가격: {price2*1.1}")
price3 = 7000
print(f"부가세 포함 가격: {price3*1.1}")
```

```python
# 위 코드를 처리하는 함수 선언
def IncludeTax (price):
    print(f"부가세 포함 가격: {price*1.1}")
    return None                             # 반환값 없어도 return None(None은 조건문에서 쓰임.)
                                            # return 기능: 반환, 함수 종료
                                            # 반환값 없으면 return만 적어도o
```

```python
IncludeTax(10000)
IncludeTax(25000)
IncludeTax(7000)
```
