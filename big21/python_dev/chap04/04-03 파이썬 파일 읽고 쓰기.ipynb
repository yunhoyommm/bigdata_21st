```python
# disk에서 파일 찾으려면 경로(PATH) 알아야 함.
# 경로 다루는 법 2가지(위치 지정 방법)
#   1. 상대 경로: 주변 위치 기반으로 경로 탐색(경로 값 바뀔 수o)
#   2. 절대 경로: 처음부터 시작(경로 값 변하지 x)

```

```python
# 파일 종류
#   1. 바이너리 파일(숫자-2진수 저장)
#     - 인코딩, 디코딩 x
#     - 이미지, 동영상, 음성, 엑셀 등(텍스트 제외한 모든 파일)
#   2. 텍스트 파일(문자 저장(사실은 숫자-2진수 저장))
#     - 메모장에 문자 저장하면 다른 형태로 저장(인코딩)
#     - 메모장 열면 문자 보임(디코딩)

# 1. 텍스트 모드, 바이너리 모드 중 결정
# 2. 읽기(R), 덮어쓰기(W), 추가쓰기(A (append)) 모드 결정
#   - 덮어쓰기 : 해당 파일 있는 경우(덮어쓰기)와 없는 경우(생성)가 있음.
```

```python
# 파일 생성: 텍스트 모드, w, 해당 파일 없는 경우(생성)
# 텍스트 파일 생성
# 파일명: "새파일", 확장자: txt, 위치: chap04
# 순서
#   1. open
#   2. write
#   3. close

# 상대경로
f = open(           # 함수
    '새파일.txt',   # file 매개변서 값 지정. 경로(경로 없으면 현재 디렉토리)와 파일명.확장자(file: FileDescriptorOrPath)
    'w'             # 텍스트, 쓰기 모드(mode: OpenTextMode = "r")             
)
f.close()           # 메소드

# 절대경로(맥과 윈도우 경로가 달라서 에러날 가능성.) 상대경로 많이 씀.
# f = open(           
#     'C:\big21\python_dev\chap04\새파일.txt',
#     'w'
# )
```

```python
# 파일을 쓰기 모드로 열어 출력값 적기
f = open("./새파일.txt", "w", encoding="UTF-8") # positional arguments , encoding은 Keyword arguments 
                                                # 파일 한 번 UTP-8이면 평생 UTF-8. 변환 불가
# 10줄을 파일에 쓰기
for i in range(1,11):
    data = f"{i}번째 줄입니다.\n" # 파일에 쓰여질 데이터 생성
    f.write(data)
f.close()
```

```python
# 파일 읽기 : 파일이 존재해야. 없으면 Error
f = open("./새파일.txt", "r", encoding="UTF-8")   # 읽을 수 있는 오브젝트 생성

line = f.readline()       # f.read: 전체(str로 나옴). f.readline: 한줄씩. f.readlines: 한줄씩 리스트로
print(line)

f.close()
```

```python
# 여러 줄 읽어서 출력
f = open("./새파일.txt", "r", encoding="UTF-8")

for i in range(10):     # 몇 줄인지 알면 for, 모르면 while
    line = f.readline() # 자주 안 쓰지만, 캐싱할 때 쪼개서 데이터 넣는 용도로 사용
    print(line)
f.close()
```

```python
# readlines() : 한 줄씩 읽고 전체를 List로 반환(제일 자주 씀)
# ["1번째 줄", "2번째 줄",...]
f = open("./새파일.txt", "r", encoding="UTF-8")
lines = f.readlines()

print(lines)
for line in lines:
    print(line)
f.close()
```

```python
# read() : 파일 내용 전체를 문자열로 반환
f = open("./새파일.txt", "r", encoding="UTF-8")
data = f.read()
print(data)
f.close()
data.split("\n")                    # 문자열 -> List
```

```python
# 파일 에 새로운 내용 추가 : 모드 -> "a"
f = open("./새파일.txt", "a", encoding="UTF-8")
for i in range(11,20):
    data = f"{i}번째 줄입니다.\n"
    f.write(data)
f.close()
```

```python
# 자동 close()
# 파일 자동 생성: foo.txt
f = open("./foo.txt", "w", encoding="UTF-8")
f.write("Life is too short")
f.close()
```

```python
with open("./foo.txt", "w", encoding="UTF-8") as f:
    f.write("Life Is Too Short")
# 자동으로 close 해줌
```

```python
# 함수: 특정 작업을 수행하는 코드 묶음에 이름을 붙여, 필요할 때마다 호출하여 사용하는 묶음
# CSV, Json 파일에 대한 처리.
# csv 파일 읽기, 쓰기하는 기능(함수)을 가지고 있는 모듈: csv 모듈 => csv.py
# 설치는 되어 있는데 외장
import csv # csv 모듈 가져오라는 명령. 외장함수라 이 명령 필요

f = open("./data/data.csv","r", encoding="UTF-8")
rdr = csv.reader(f)     # csv 형태로 읽을 수 있는 구조 생성
for line in rdr:
    print(line)
f.close()
```

```python
# CSV 파일의 쓰기
import csv
# data 디렉토리에 output.csv 파일 생성하고 쓰기
f = open("./data/output.csv", "w", encoding="UTF-8")
# csv write 할 수 있는 오브젝트 생성
wr = csv.writer(f)
wr.writerow([1,"홍길동",False])
wr.writerow([2,"고길동",True])
wr.writerows([[3,"마이콜",True], [4,"이길동",True]])
f.close()
```

```python
# JSON 파일 입출력
# 자바스크립트에서 오브젝트를 표기하는 문법: {key:value, key:value,...} <- python에서는 딕셔너리
# JSON(파일) -> DICT ->JSON(파일) : JSON module -> function : load(), dump()
```

```python
# JSON 쓰기: python 객체(dict) -> json 형태의 파일로 저장
import json

data = {            # 딕셔너리 생성
    "이름":"홍길동",
    "나이":30,
    "취미":["코딩","운동"]
}
# json 형식으로 파일에 저장
with open("./data/profile.json","w",encoding="UTF-8") as f:
    json.dump(              # dict -> json 저장
        data,               # 파일에 저장할 딕셔너리 지정
        f,                  # 어느 파일?
        ensure_ascii=False, # 한글 깨지지 않고 그대로 저장하기 위해
        indent=2            # 2칸 들여쓰기
    )
```

```python
# json 형식의 파일 -> 파이썬 객체(dict) 읽기
with open("./data/profile.json","r",encoding="UTF-8") as f:
    dict_data = json.load(f)
print(dict_data.get("이름"))
print(dict_data.get("나이"))
print(dict_data.get("취미"))
```
