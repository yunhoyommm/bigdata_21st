```python
# 조건문 : if, switch, match ~ case
# 기본은 if 사용. 복잡해지면 switch or match 고려
# switch, match에는 제약사항 존재
# if문의 3가지 구조
# 1. if 구조 : 특정 조건에 만족할 경우에만 특정 코드 실행
#   if 조건식 :
#       조건식이 True일 때 실행될 코드 블록
# 2. if ~ else 구조
#   if 조건식 :
#       조건식이 True일 때 실행될 코드 블록
#   else :
#       조건식이 False일 때 실행될 코드 블록
# 3. if ~ elseif ~ else 구조
```

```python
# "돈이 있으면 택시를 타고, 돈이 없으면 걸어간다"
# 변수명 선언 : 영어, 카멜표기법 사용(바이브코딩할 때 표기법 설정)(예. studentCount)
# 돈 저장할 변수 필요 : money 0, 1
money = 0
if money :                      # :(콜론)의 기능 -> 블록의 시작
    print("택시를 타고 가라")    # 들여쓰기가 블록의 기능
else :                          # False일 때 블록의 시작
    print("걸어가라")
```

```python
x = 3
y = 5
print(x == y)  # 비교연산자 결과는 boolean type : True or False
if x == y:
    print("같다")
else:
    print("같지 않다")
```

```python
# "만약 5000원 이상의 돈을 가지고 있으면 택시 타고, 그렇지 않으면 걸어와라"
money = 5000
if money >= 5000 :
    print("택시 타고 가라")
else :
    print("걸어가라")
```

```python
# "돈이 5000원 이상 있거나 카드가 있으면 택시를 타고, 그렇지 않으면 걸어와라"
money = 4000    # 돈은 4000원을 가지고 있다.
card = 1        # 카드를 가지고 있다.
if money >= 5000 or card :  # 돈을 5000원 이상 가지고 있거나, 카드 있거나
    print("택시를 타고 가라")
else :
    print("걸어가라")
```

```python
# in 연산자 => 특정 값 in iterable(반복가능한 객체) 사용
# iterable 
# - 여러 개 저장하고 있는 객체. 
# - 반복 가능한 객체.
# - 인덱스를 가지고 있는 객체
# list, tuple, string <= iterable 객체
print(1 in [1,2,3]) # 1이 리스트 [1,2,3] 안에 있니?
print(12 in [1,2,3]) # 1이 리스트 [1,2,3] 안에 있니?
```

```python
# "주머니에 돈이 있으면 택시를 타고, 없으면 걸어가라"
pocket = ["paper","cellphone","money"]
if "money" in pocket :  # 주머니 안에 돈이 있으면 
    print("택시를 타고 가라")
else :
    print("걸어가라")
```
