```python
# 딕셔너리: 데이터 타입 -> 저장소 : 속성(변수)
# 데이터 타입 별로 저장하는 방식과 종류, 형태가 다름.
# 1. 숫자형: 정수, 실수 저장소
# 2. 문자열
#   종류: 문자열 저장소. 
#   형태: 인덱스
#   특징: 불변
# 3. 리스트
#   종류: 모든 타입 저장 가능
#   형태: 인덱스 사용
#   특징: 가변
# 4. 튜플
#   종류: 모든 타입 저장 가능
#   형태: 인덱스 사용
#   특징: 불변
# 5. 딕셔너리
#   종류: key - 문자열 권장, value - 모든 타입 저장 가능
#   형태: 인덱스 x. key 사용.
#   특징: 가변
# 6. 집합(셋)
#   종류: 모든 타입 저장 가능
#   형태: 인덱스 x, key x. 중복 허용

# 인덱스 : 0부터 시작하고 1씩 증가하는 값 (규칙 존재)
#         저장소 접근할 때 사용
# 인덱스 사용 예 : names = ["홍길동","고길동","마이콜"]
# , kors = [80,90,60]
# name[0] <- "홍길동"   # 리스트 생성과 활용이 붙어있으면 쉽게 알 수 o
# kors[2]               # 누구의 점수인지 한 번에 알 수 없음
```

```python
# 1. 딕셔너리 생성 방법 : {키1:값1, 키2:값2}
# 국어점수
names = ["홍길동","고길동","마이콜"]
kors = [80,90,60]
print(kors[2])


# 2. 딕셔너리 처리
kors_dict = {
    "홍길동":80,            # key : 홍길동, value : 80
    "고길동":90,
    "마이콜":60
}
print(kors_dict["마이콜"])  # "마이콜" key로 마이콜의 국어 점수 접근
```

```python
# 1. 딕셔너리 쌍(key:value) 추가하기
a = {1:"a"}     # 딕셔너리 만든 이유는 가독성인데 숫자면 인덱스랑 다른 게 없..
print(a[1])     # 키로 숫자 쓰는 건 좀...
# 추가 2:"b"
a[2] = "b"
print(a)
print(a[2])
# 추가 : name:"pey"
a["name"] = "pey"
print(a)
print(a["name"])
# 추가 리스트 : 3:[1,2,3]
a[3] = [1,2,3]
print(a)
```

```python
# 2. 딕셔너리 value 삭제
del a[1]    # 잘 쓰지 않음. 메서드 주로 활용

print(a)
```

```python
# 메소드(method) ⊂ 함수(function)
# Function : 입력(재료)받아 처리(기능)하고 반환(return)
# 함수: 함수는 데이터를 넣으면 규칙(코드)에 따라 결과물을 반환
# 함수의 구조(예. 자판기)
#   - 입력값: 자판기에 넣는 돈
#   - 함수 내부(코드): 자판기 안에서 음료수 떨어뜨리는 기계
#   - 반환값: 자판기 툭 떨어지는 음료수

# 함수를 왜 사용하는데? : 코드의 반복을 줄이기 위해 사용
```

#### 메소드 vs 함수
|구분|함수(Function)|메소드(Method)|
|----|----|----|
|소속|독립적 존재(특정 객체에 종속x)|오브젝트에 종속(클래스 내부에 정의)|
|호출방법|함수이름() 형태로 직접 호출|오브젝트이름.메소드이름() 형태로 오브젝트 통해 호출|

```python
# 딕셔너리 메소드: 암기해야 함.
# 1. keys() : 키들을 반환
a = {
    'name':'pey',
    'phone':'01012341234',
    'birth':'1118'
}
print(a)
# keys_a = a.keys()             # keys_a는 dict_keys
keys_a = list(a.keys())         # list로 변환해야 인덱싱 가능
print(keys_a[0], type(keys_a))
```

```python
# values 추출 메소드 : values() => 값들을 dict_values 오브젝트에 담아서 변환
# b = list(a.values())    # 리스트로 변환
# b[0]

list(a.values())[0]
```

```python
# key와 value 한 쌍으로 추출 메소드: items()
print(a)
print(a.items())
```
