```python
""" 여러 줄 주석도 가능
# 오른쪽 상단에 파이썬 버전 떠야 에러x

# 문자열 데이터 타입: 여러 문자를 저장하는 공간 생성 -> '', ""
# 문자 != 문자열
# 문자: 한 문자
# 문자열: 여러 문자
# 파이썬은 문자도 문자열로 처리

"""
a = "Hello World"   # 변수는 계속 쓰려고 만드는 것
print(a)
```

```python
# 1. 문자열 안에 작은 따옴표나 큰 따옴표 포함시키고 싶을 때
# a = 'Python's favorite food is perl' Error
a = "Python's favorite food is perl"
print(a)
b = '"Python is very easy." he says!'
print(b)
```

```python
# 여러 줄인 문자열을 변수에 대입하고 싶을 때
multiline = "Life is\nto short\nYou need\n python"
print(multiline)        # 이스케이프 문자는 print의 기능
multiline               # jupyter는 변수만 입력해도 출력 가능
                        # 텍스트 데이터는 원래 1줄. 기능을 통해 여러 줄로 나뉜 것.
```
