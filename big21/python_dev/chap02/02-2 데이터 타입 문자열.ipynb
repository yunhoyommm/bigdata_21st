```python
# 문자열 인덱싱 & 슬라이싱 : 매우 중요
# 1) 인덱싱: 단일 값 접근(추출, 대입 위해). [] 연산자 사용
# 역방향도 씀. -로 반대 방향 접근
# 역방향 인덱스 활용 예시: 1600자 문장의 마지막 글자 뽑아오기. 정방향은 글자 수를 알아야..
a = "abcdefg"
print(a)
print(a[2])
print(a[-5])

#'f' 출력
print(a[-2])
```

```python
# 2) 슬라이싱: 여러 값 접근(추출, 대입). [] 연산자 안에 : 을 사용
a = 'abcdefg'
# "bcd" 추출
print(a[1]+a[2]+a[3])
print(a[1:3])
print(a[1:4])   # [start index : end index-1] 추출

# "efg" 추출(끝까지 추출)
print(a[4:7])   # 권장x. 7 OutOfBounds인데 정상작동. but 7이 어느 조건에서는 오류가 되기도 함.
print(a[4:])    # 권장

# "fg" 추출
print(a[5:])
print(a[-2:])   # 처음 값 찾는 건 -지만 끝까지는 정방향.
```

```python
# 슬라이싱 사례 정리
# [시작값:끝값] : 시작값부터 끝값-1까지 추출
# [시작값:끝값x] : 시작값부터 끝까지 추출
# [시작값x:끝값] : 처음부터 끝값-1까지 추출
# [시작값x:끝값x] : 처음부터 끝까지 추출

a = "abcdefg"
b = a           # b에 a의 주소값 대입
c = a[:]        # 대괄호 콜론 쓰면 값이 복사됨. 복사한 오브젝트의 주소에 대입
print(id(a))
print(id(b))
print(id(c))    # 문자열은 튜플 성격이라 복사가 안 됨. 그래서 메모리 주소 같음. 문자열은 불변 오브젝트. 문자열은 상수
```

```python
data = "20010331Rainy"
# 연도 추출: 2001
print(data[:4])
# 월,일 추출: 0331
print(data[4:8])
# 날씨 추출
print(data[8:])
```

```python
# 포맷 코드와 숫자 함께 사용
# 1) 정렬과 공백
a = "%10s" % "hi"       # %10s: 문자열 10자리 공간 만들기 (오른쪽 정렬)
b = "%-10s" % "hi"      # %-10s: 왼쪽 정렬
print(a)                # +10이면 마지막 2자리에 채워짐.
print(b)
print(len(a))           # len(iterable object) 반복가능한 오브젝트 (대표 예: 문자열) 길이 반환
print(len(b))
```

```python
# 소수점 표현
"%10.4f" % 3.43127234     # 전체 문자를 10자리로 만들어라. 그 중 소수점을 4자리로 써라. (예. 3.4312면 4자리 남아있음)
"%0.4f" % 3.43127234      # 반올림 가능(0은 공간을 맞춰서 만들겠다는 뜻)

```

```python
# 문자열 오브젝트 : 문자열(값) + method로 구성
# 해당 오브젝트의 문자열을 처리하는 특정 기능
a = "hobby"           # 문자열 오브젝트를 생성하라는 명령어
print(a.count('bb'))   # a가 가리키는 문자열에서 'b' 라는 문자의 개수 반환하는 메소드
```

```python
# 위치 알려주기1 : find(찾을 문자열) -> 인덱스 값 반환
# 예1. a = "abcdefg" -> a.find('a') -> 0 (대체로 인덱싱, 슬라이싱을 위해 사용)
# 예2. a = "abcaefg" -> a.find('a') -> 0 (해당 문자열 여러 개일 경우 첫 번째 a 위치 반환.)
# 예3. a = "abcaefg" -> a.find('k') -> -1 (찾는 문자열이 없으면 -1 반환)

a = "Python is best choice"
# 'b' 찾기
# a.find("best")      # "best"가 시작하는 위치
a.find("not")         # "not" 없으니 -1 출력
```

```python
# 위치 알려주기2 : index(찾을 문자열) : 인덱스 정수를 반환
# find()와의 차이: 해당 문자열 없는 경우 find()는 -1, index()는 ValueError 예외 발생(예외 처리 필요)
a = "Life is too short"
# a.index("not")
a.find("t")
```

```python
# 문자열 삽입 : join("문자열"). 문자열 중간에 특정 문자 삽입
# 예. 원래 "abcd" 문자열이 있다 -> a,b,c,d
a = "/"
b = a.join("abcd")
print(b)
```

```python
# 소문자 -> 대문자: upper()
# 대문자 -> 소문자: lower()
# 인코딩: 문자(A) -> 아스키코드(A->65) -> 숫자로 변경(65)
# 디코딩: 숫자 -> 문자
a = "hi"
b = a.upper()
print(b)

c = b.lower()
print(c)
```

```python
# 공백 지우기(중간 공백은 삭제 불가 -> replace() 사용 필요)
# lstrip(): 문자열 왼쪽 공백 삭제
# rstrip(): 문자열 오른쪽 공백 삭제
# strip(): 문자열 양쪽 공백 삭제

a = "  hi hi  "
print(a.strip())
```

```python
# 문자열 내용 변경: replace("찾을 문자열", "변경할 문자열")
a = "Life is too short"
print(a)
b = a.replace("Life", "Your leg")
print(b)
```

```python
# 문자열 나누기 : split() -> 리스트 반환
# split(x): 공백 기준으로 나누기 진행
# split(특정 문자 패턴 명시): 문자 따라 나누기 진행

a = "Life is too short" # 문자열 오브젝트 생성. a라는 변수에 대입. a로 접근 가능
print(a.split())        # object 있어야 method 사용 가능

b = "a:b:c:d"
print(b.split(":"))
```

```python
# 고급 문자열 포매팅
# 포매팅 3가지 방법
# 1. % 사용: 이제 사용x
# 2. format() 메소드 사용: 오늘 설명
# 3. f-string 사용: 다양하게 사용. 3가지 중 가장 많이 사용.
```

```python
# 1 % 사용
num = 10
a = "I eat %d apples" % num
print(a)
# 2. format()
b = "I eat {} apples"
print(b.format(num))
# 3. f-string
print(f"I eat {num} apples")    # 속도도 가장 빠름 & 가독성
```

```python
# f-string
# 정렬: 왼쪽, 오른쪽, 가운데
h = "Hello"
print(h)
print(f"{h}")
print(f"{h:10}")    # 공간 10자리 생성 후 왼쪽 정렬
print(f"{h:>10}")   # 공간 10자리 생성 후 오른쪽 정렬
print(f"{h:^10}")   # 공간 10자리 생성 후 가운데 정렬
```

```python
# f-string에서 {}는 표현식이라 부름. 중괄호 안에 식(코드)이 들어갈 수 있다.
# f-string에 조건식 부여 : if 문
h = "hello"
b = "bye"

print(f"{h if len(h) > len(b) else b}")
print(f"{h if len(h) < len(b) else b}")
```

```python
# 연산 가능
a = 5
b = 3
result = a * b
print("a = 5, b = 3 a * b = 15")
print(f"a = {a}, b = {b} a * b = {result}")
print(f"a = {a}, b = {b} a * b = {a*b}")
```
