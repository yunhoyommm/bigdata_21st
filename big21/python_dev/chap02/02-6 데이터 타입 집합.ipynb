```python
# set([]) 함수 사용, 여러 개 저장
s1 = set([4,1,3,2,2,3,4]) # 집합 오브젝트 생성, 중복 제거하고 생성
print(s1)
print(list(s1)[0])            # print(s1[0])은 인덱싱되지 않음.(set은 인덱스가 없음)

# 예. 1000개 팔린 물품의 종류가 궁금하면 set 돌려서 종류 확인
```

```python
s2 = set('Hello')
print(s2)
```
