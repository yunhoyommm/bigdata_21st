```python
# set([]) 함수 사용, 여러 개 저장
s1 = set([4,1,3,2,2,3,4]) # 집합 오브젝트 생성, 중복 제거하고 생성
print(s1)
print(list(s1)[0])            # print(s1[0])은 인덱싱되지 않음.(set은 인덱스가 없음)

# 예. 1000개 팔린 물품의 종류가 궁금하면 set 돌려서 종류 확인
```

```python
s2 = set('Hello')
print(s2)
```

```python
# 집합(set) : 중복 제거, 합집합, 교집합, 차집합
s1 = set([1,2,3,4,5,6])
s2 = set([4,5,6,7,8,9])
# 교집합
# s1 & s2           # & 기호 사용
s1.intersection(s2) # method
# 합집합
# s1 | s2           # | 기호 사용
s1.union(s2)        # method
# 차집합
# s1 - s2           # - 기호 사용
s1.difference(s2)
```

```python
# 집합 메소드
# add(추가), update(수정), remove(제거)
# update : 사실은 여러 개 추가
s1 = set([1,2,3])   # [1,2,3] 리스트가 set으로 변환
s1.update([4,5,6])  # [4,5,6], (4,5,6) 어느 것이든 set으로 변환됨.
print(s1)
```

```python
s1 = set([1,2,3])
s1.update([4])    #TypeError: 'int' object is not iterable(단일 값은 안 됨.)
# 중요!
# update(iterable)
# iterable class : 반복 가능한 클래스(인덱스가 존재. 인덱싱, 슬라이싱 가능)
# 리스트는 iterable을 상속받은 오브젝트 : ""
# 튜플은 리스트를 상속받은 오브젝트
# 문자열은 튜플을 상속받은 오브젝트

print(s1)
```
