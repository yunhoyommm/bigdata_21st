```python
# : 한 줄 주석( ctrl + / )
# 파이썬 자료형 숫자형 : 
# 자료형: Data Type. 저장 공간을 생성하라는 명령. 이름(변수명) 부여
# 숫자 자료형: 정수(int) & 실수(소수점. float)
print(type(10))
print(type(10.))

# python은 동적할당. '10' -> '=' -> 'int_num' 순으로 진행.
# 동적할당: 대입될 때 변수가 생성.
# num = 10            # 이 때 num은 정수 공간 부여된 변수명
# print(type(num))
# num = 10.           # num 실수 공간에 부여된 변수명
# print(type(num))


num = 10
num = '10'

print(num+str(1))
# print(type(num))
```

```python
# 정수 데이터 타입
# 정수: 음의 정수, 0, 양의 정수
a = 123     # 123의 주소값을 a에 대입. 목적: a로 123 정수를 사용
a = -123    # -123의 주소값을 a에 대입. 
a = 0       
```

```python
# 실수 데이터 타입
a = 1.2
a = -3.45
a = 0.0
a = 42400000000.0
a = 4.24E10         # 소수점을 10자리 뒤로 넘겨라.
a = 4.24E2          # 4.24 -> 424.0
a = 4.24e-2         # 4.24 -> 0.0424
```

```python
# 16진수: 0~15까지의 값으로 숫자를 표기
# 한 자리 값으로 표기: 0 ~ 9, A ~ F
# 표기법: '0x'로 시작
a = 0xAB    # 16진수 값 AB
# 2진수 표기했을 때 길어지므로 16진수 사용. 유니코드에서 16진수 사용
a = 171
print(a)
```

```python
# 연산자
# 사칙연산(+, -, *, /)
a = 3 # a에 3을 대입
b = 4
print(a + b)
print(a / b)    # 실수 출력(0.75)
```

```python
# 거듭제곱 연산자 : **
a = 3
b = 3
print(a**b) # 3 ^ 3
```

```python
# 나머지 구하는 연산자: %
# 홀짝, 배수 구할 때 사용
a = 7
b = 2
print(a % b)
```

```python
# 나눗셈 후 소수점 아랫 자리를 버리는 연산자 : //
# 나누기
a = 7
b = 4
print(a / b)
print(a // b)   # 두 정수(1, 2) 중 작은 값 반환(내림)

a = -7
print(a / b)
print(a // b)   # -1이 아닌 -2가 출력
                # 두 정수(-2, -1) 중 작은 값 반환(내림)
```
