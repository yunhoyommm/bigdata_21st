```python
# 조건식 자리에 데이터 타입이 들어가면 참/거짓의 역할을 한다.
# 데이터 타입(변수)에 값이 존재하면 참, 없으면(비어있으면) 거짓 역할.

# 숫자
# 기본 if
# if 조건식 자리:
#   조건식이 True 일 때 실행될 코드가 들어갈 자리
# else :
#   조건식이 False 일 때 실행될 코드가 들어갈 자리


num = 10
if num :
    print("num의 값은 0이 아니다")
else : 
    print("num의 값은 0이다")
```

```python
list_ = [1]  # 빈 리스트
if list_ :
    print("빈 리스트가 아니다")
else :
    print("빈 리스트")
```

```python
a = [1,2,3,4]
# while 조건식 자리. => 현재 상태: 조건식 자리에 리스트 
# 조건식 True일 때 아래 코드를 실행하는 것이 while
while a:            
    print(a.pop())
```
