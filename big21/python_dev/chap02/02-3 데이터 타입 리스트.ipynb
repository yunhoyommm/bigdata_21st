### 파이썬 자료형 리스트
#### 리스트
- 여러 개 저장
- 여러 개를 하나로 묶을 때 사용
---

```python
# 학생 3명.
# 이름(문자열) 저장
name1 = "홍길동"
name2 = "고길동"
name3 = "마이콜"
# 리스트
names = ["홍길동", "고길동", "마이콜"]

print(names)
```

```python
# 리스트 생성, 사용
# 1~10 사이의 홀수 저장: 1,3,5,7,9
# 리스트 생성 : [] 대괄호, list()
odds = [1,3,5,7,9]
print(odds)
# 3 출력
print(odds[1])
```

```python
# 어떤 타입을 저장할 수 있는지
a = []                              # 빈 리스트, 빈 값을 저장. a = list()와 동일
b = [1,2,3]                         # 정수 저장
c = ["Life", "is", "too", "short"]  # 여러 개의 문자열을 한 번에 저장
d = [1,2,"Life","is"]               # 타입이 달라도 저장 가능
e = [1,2,["Life","is"]]             # 리스트 안에 리스트 저장
```

```python
# 리스트의 인덱싱, 슬라이싱
# 인덱싱 : 단일 값 추출. []
# 슬라이싱 : 여러 값 추출. [처음값:끝값-1]
a = [1,2,3]
print(a)
# 3 추출
print(a[2])
print(a[-1])
```

```python
#
a = [1,2,3,['a','b','c']]
print(a[3][2])
print(a[1])
print(a[3])
print(a[-1])
```

```python
# 슬라이싱
a = [1,2,3,4,5]
# 1, 2 뽑기
print(a[:2])
# 3,4,5 뽑기
print(a[2:])
```

```python
a = [1,2,3,4,5]
b = a
c = a[:]        # 복사된 슬라이싱 리스트는 다른 주소에 담김
print(id(a), id(b), id(c))
```

```python
# 어떤 반에 3명의 학생이 있다. 이름과 국어 점수 저장 출력
students = [
    ["홍길동","고길동","마이콜"],     # 이름 저장하는 리스트
    [80,90,60]                      # 점수 저장하는 리스트
]
print(students)
print(students[1][1])               # 고길동의 국어 점수
print(students[0][1])
```

```python
# 리스트 연산자
a = [1,2,3]
b = [4,5,6]
# 두 개 리스트 합치기(+) : 사용x
print(a + b)

# 리스트 반복(*)
print(a * 3)
```

```python
# 리스트의 수정, 변경, 삭제
# 1. 하나의 값(인덱싱) 수정
a = [1,2,3]
a[2] = 4
print(a)
```

```python
# 2. 리스트에서 연속된 범위의 값(슬라이싱) 수정
# 현재 a = [1,2,4]
# print(a[1:2])             # 결과: [2] (리스트의 2 출력)
# print(a[1])               # 결과: 2 (값 2 출력)
a[1:2] = ["a","b","c"]      # 결과: [1, 'a', 'b', 'c', 4]
# a[1] = ["a","b","c"]      # 결과: [1, ['a', 'b', 'c'], 4]
print(a)
```

```python
# 리스트 요소 삭제
a = [1, 'a', 'b', 'c', 4]
a[1:3] = []         # 양이 많을 경우 이 방법이 어려워서 권장x & 가독성 낮음.
print(a)
```

```python
# 리스트 요소(object) 삭제 : del 함수 사용
# del 함수는 메모리에서 모든 오브젝트 삭제 가능
del a[1]
print(a)
```

```python
# 리스트 관련 메소드
# 1. 리스트 마지막에 추가 : append(추가할 값)
a = [1,2,3]
# 맨 뒤에 4 추가
a.append(4)
print(a)
# 현재 리스트 안에 리스트 추가(append) : [1, 2, 3, 4, [5,6]]
a.append([5,6])
print(a)
```

```python
# 리스트 정렬 : sort() -> 오름차순(1,2,3,4) 정렬만. 내림차순(4,3,2,1)
a = [1,4,3,2]
a.sort()
print(a)        # 오름차순
print(a[::-1])  # 내림차순 
# 리스트 증감값 있음. [start : end : step]
```

```python
# 알파벳 순서 정렬: 오름차순만 가능
a = ["a","c","b"]
a.sort()
print(a)
```

```python
# 리스트 뒤집기 : reverse(). 정렬은 아님. 그냥 뒤집기
a = ["a","c","b"]
a.reverse()
print(a)
```

```python
# 위치 반환 : index(찾을 값)
# 찾을 값 있을 경우: 해당 위치의 인덱스 값 반환
# 찾을 값 없을 경우: 예외 발생
a = [1,2,3]
print(a.index(3))     # 2: 인덱스 값 반환
# print(a.index(5))   # ValueError: 4 is not in list
```

```python
# 리스트에 요소 삽입 : insert(삽입할 위치, 삽입할 값)
a = [1,2,3]
# 첫 번째 자리에 4 삽입
a.insert(0, 4)
print(a)

# 네 번째 자리에 5 삽입
a.insert(3, 5)
print(a)
```

```python
# 리스트 요소 제거: remove(제거할 값)
a = [1,2,3,1,2,3]
# 3 제거
a.remove(3)
print(a)
a.remove(3)
print(a)
```

```python
# 리스트에서 특정 값 추출 
# pop(): 마지막 값을 끄집어낸다(자주 씀)
a = [1,2,3]
b = a.pop()
print(a, b)
# pop(특정 인덱스): 특정 위치의 값을 끄집어낸다(잘 안 씀)
c = a.pop(1)
print(a, c)
```

```python
# 리스트에 포함된 값의 개수 세기 : count(찾을 값)
a = [1,2,3,1]
a.count(1)
```

```python
# 리스트 확장 : extend(리스트)
a = [1,2,3]
b = [4,5,6]
# print(a.extend(b))  # return 값 None
# print(a)            # [1, 2, 3, 4, 5, 6]
b.extend(a)
print(a)
print(b)            # [4, 5, 6, 1, 2, 3]
```
