```python
# 튜플 vs 리스트 공통점 : 인덱스 존재 => 인덱싱, 슬라이싱 가능
# 불변 vs 가변(중요!)
# 튜플 생성
t1 = ()         # 빈 소괄호 => 빈 튜플 오브젝트 생성
t2 = (1,)       # 1이 들어 있는 튜플 오브젝트 생성
t3 = (1,2,3)    # t3 = 1,2,3 과 동일.(중요!) 이 부분 꼭 기억하기.
```

```python
# 불변 확인 : 삭제, 수정해 보기 -> 삭제, 수정 불가
t1 = 1,2,3,4
# del t1[0]    # TypeError: 'tuple' object doesn't support item deletion
# t1[0] = 5    # TypeError: 'tuple' object does not support item assignment
```

```python
# 튜플은 인덱스를 가지고 있다.(인덱싱, 슬라이싱 가능)
# 튜플 더하기
t1 = (1,2,"a","b")
t2 = (3,4)
print(t1 + t2)
print(t1)
print(t2)
```

```python
# 정렬 알고리즘
a = 5
b = 3
print(a, b)
a, b = b, a # 튜플 형태.
print(a, b)
```
