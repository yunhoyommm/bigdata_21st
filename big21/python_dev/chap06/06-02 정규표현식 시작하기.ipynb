```python
# 정규식 이용한 문자열 검색
# compile : 같은 패턴을 반복 사용할 경우 컴파일.
import re   # regex 모듈(re에서 확장된 모듈)

pat_str = '[a-z]+'  # 알파벳 소문자 1자 이상 패턴
pat = re.compile(pat_str)   # 패턴 object 생성 -> 메소드 사용 가능(match(), search(), findall() 등)
m1 = pat.match('python')     # object 안의 메소드(반복 사용 가능)  통과하면 Match object return, 통과 x 면 None
# re.match()      # function(1번만 사용 가능)
print(m1)   # <re.Match object; span=(0, 6), match='python'>
m2 = pat.match("3 python")  # 첫 글자부터 찾는데 3으로 시작.(소문자 x) => None
print(m2)   # None
m3 = pat.search("3 python") # 전체 대상으로 찾음(중간부터도 가능)
print(m3)   # <re.Match object; span=(2, 8), match='python'>
```

```python
# 언제 사용?
# match: 이 줄이 패턴으로 시작하니?
# search: 이 안에 패턴이 있니?
# findall: 전부 찾아줘.
ptr_str = "[a-z]+"  # 한 글자 단어 찾기
pat = re.compile(pat_str)
result = pat.findall("life is too short")   # 4글자/2글자/3글자/5글자
print(result) # ['life', 'is', 'too', 'short']
```

```python
ptr_str = '[a-z]+'
pat = re.compile(ptr_str)
result = pat.finditer("life is too short")
print(result)   # iterable 객체 출력
for res in result:
    print(res)
```

```python
# Match object, method 설명
# group(), span(start, end)-> start(), end()
# group()
m = pat.search("3 python")
print(m.span())     # (2, 8)  -> (p, n)
print(m.start())    # 2
print(m.end())      # 8
print(m.group())    # python
```

```python
import re

pat = re.compile(r"\bclass\b")
print(pat.search("no class at all"))
print(pat.search("class at all"))
print(pat.search("no subclass at all"))

pat = re.compile(r"\Bclass\B")
print(pat.search("no subclassaaa at all"))  # "subclassaaa"가 아니라 "class"만 찾아줌.
```

```python
# 1. class 가 포함된 모든 단어 찾기
pat_str = r'\b\w*class\w*\b'
pat = re.compile(pat_str)
print(pat.search('no subclassaaa at all'))
print(pat.search('no class at all'))
```

```python
# 1. class로 시작하는 모든 단어 찾기
pat_str = r'\bclass\w*\b'   # class로 시작하는 단어 찾기
pat = re.compile(pat_str)
print(pat.search('no subclassaaa at all'))
print(pat.search('no classes at all'))
```

```python
import re

pat_str = r"(\w+)\s+(\d+[-]\d+[-]\d+)"
pat = re.compile(pat_str)
m = pat.search("park 010-1234-1234")
print(m.group(0))   # 전체
print(m.group(1))
print(m.group(2))
```

```python
# 그룹핑된 문자열 재참조하기
p = re.compile(r"(\b\w+)\s+\1")     # \b(경계)\w(알파벳)+(1자이상)\s(공백)   # \1: (\b\w+) 중복된 것 찾아 
p.search("Paris is the the spring").group()
```

```python
import re
pat_str = r"(?P<name>\w+)\s+(?P<phone>\d+[-]\d+[-]\d+)"
pat = re.compile(pat_str)
m = pat.search("park 010-1234-1234")
print(m.group("name"))
print(m.group("phone"))
```
